```python
print(fd["InGamePurchases"].value_counts())  ## Printint purchase stats

sns.set(style="whitegrid")  ## Setting style for pie chart

labels = ["No Purchases", "Purchases"]
data = fd["InGamePurchases"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data,
        labels=labels,
        autopct="%1.2f%%",
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05, 0.05])
plt.title("In-Game Purchases", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
for col in ["Gender", "Location", "GameGenre", "EngagementLevel"]:
    print(f"\n{col} vs. In-Game Purchases:")
    print(pd.crosstab(fd[col], fd["InGamePurchases"]))

    col_freq = fd[col].value_counts()   ## Getting total number of values of metric for each column
    sns.set(style="whitegrid")

    plt.figure(figsize=(8, 5))
    sns.barplot(x=col_freq.index, y=col_freq.values, palette='viridis')

    plt.title(f'Frequency of {col} Distribution', fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.xticks(ticks=range(len(col_freq.index)), labels=[str(level) for level in col_freq.index], rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
## Plotting Engagement level distribution for each class of engagement
print("\nEngagement Level Distribution:")
print(fd["EngagementLevel"].value_counts())

plt.figure(figsize=(8, 6))
engagement_level_freq = fd["EngagementLevel"].value_counts()
sns.set(style="whitegrid")
sns.barplot(x=engagement_level_freq.index, y=engagement_level_freq.values, palette="viridis")

plt.title("Engagement Level Distribution", fontsize=16)
plt.xlabel("Engagement Level", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Plotting Purchases by each class of Engagement Level
purchases = fd[fd['InGamePurchases'] == 1]
print("\nEngagement Level Distribution for In-Game Purchases:")
print(purchases["EngagementLevel"].value_counts())
engagement_level_freq = purchases.groupby('EngagementLevel').size()

plt.figure(figsize=(8, 6))
sns.barplot(x=engagement_level_freq.index, y=engagement_level_freq.values, palette="magma")
plt.title('Frequency of In-Game Purchases by Engagement Level', fontsize=16)
plt.xlabel('Engagement Level', fontsize=14)
plt.ylabel('Frequency of In-Game Purchases', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Plotting Distribution of Purchases by Engagement Level
grouped_data = fd.groupby("EngagementLevel")["InGamePurchases"].value_counts().unstack()
total_purchases_by_engagement = grouped_data.sum(axis=1)

## Creating pie chart for percentage of buys for each class of engagement level
plt.figure(figsize=(6, 6))
plt.pie(total_purchases_by_engagement,
        labels=total_purchases_by_engagement.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05] * len(total_purchases_by_engagement))


plt.title("Purchases vs Engagement Level", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
## Plotting Gender distribution for each class of Gender
print("\nGender Distribution:")
print(fd["Gender"].value_counts())

plt.figure(figsize=(8, 6))
Gender_freq = fd["Gender"].value_counts()
sns.set(style="whitegrid")
sns.barplot(x=Gender_freq.index, y=Gender_freq.values, palette="viridis")

plt.title("Gender Distribution", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Plotting Purchases by each class of Gender
purchases = fd[fd['InGamePurchases'] == 1]
print("\nGender Distribution for In-Game Purchases:")
print(purchases["Gender"].value_counts())


Gender_freq = purchases.groupby('Gender').size()
plt.figure(figsize=(8, 6))
sns.barplot(x=Gender_freq.index, y=Gender_freq.values, palette="magma")

plt.title('Frequency of In-Game Purchases by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Frequency of In-Game Purchases', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Plotting Distribution of Purchases by Gender
grouped_data = fd.groupby("Gender")["InGamePurchases"].value_counts().unstack()
total_purchases_by_Gender = grouped_data.sum(axis=1)

## Creating pie chart for percentage of buys for each class of gender
plt.figure(figsize=(6, 6))
plt.pie(total_purchases_by_Gender,
        labels=total_purchases_by_Gender.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05] * len(total_purchases_by_Gender))

plt.title("Purchases vs Gender", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
## Plotting location distribution for each class of Location
print("\nLocation Distribution:")
print(fd["Location"].value_counts())

plt.figure(figsize=(8, 6))
location_freq = fd["Location"].value_counts()

sns.set(style="whitegrid")
sns.barplot(x=location_freq.index, y=location_freq.values, palette="viridis")

plt.title("Location Distribution", fontsize=16)
plt.xlabel("Location", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Plotting Purchases by each class of Location
purchases = fd[fd['InGamePurchases'] == 1]
print("\nLocation Distribution for In-Game Purchases:")
print(purchases["Location"].value_counts())

location_freq = purchases.groupby('Location').size()
plt.figure(figsize=(8, 6))
sns.barplot(x=location_freq.index, y=location_freq.values, palette="magma")

plt.title('Frequency of In-Game Purchases by Location', fontsize=16)
plt.xlabel('Location', fontsize=14)
plt.ylabel('Frequency of In-Game Purchases', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

## Creating pie chart for percentage of buys for each class of gender
grouped_data = fd.groupby("Location")["InGamePurchases"].value_counts().unstack()
total_purchases_by_location = grouped_data.sum(axis=1)

plt.figure(figsize=(6, 6))
plt.pie(total_purchases_by_location,
        labels=total_purchases_by_location.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05] * len(total_purchases_by_location))

plt.title("Purchases vs Location", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
income_avg = fd.groupby('InGamePurchases')['EstimatedAnnualIncome'].mean()  ## Calculating average income for each buy class
overall_avg_income = fd['EstimatedAnnualIncome'].mean() ## Calculating overall mean

income_data = pd.DataFrame({
    'Group': ['Overall', 'InGamePurchase_1', 'InGamePurchase_0'],
    'Avg_EstimatedAnnualIncome': [overall_avg_income, income_avg[1], income_avg[0]]
})

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Group', y='Avg_EstimatedAnnualIncome', data=income_data, palette='viridis')

plt.title('Average Estimated Annual Income by In-Game Purchases', fontsize=16)
plt.xlabel('Group', fontsize=14)
plt.ylabel('Average Estimated Annual Income', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
new_data = fd.copy()
new_data["IncomeBin"] = pd.cut(fd["EstimatedAnnualIncome"], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000])

grouped_data = new_data.groupby("IncomeBin")["InGamePurchases"].apply(lambda x: (x == 1).sum())

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x=grouped_data.index.astype(str), y=grouped_data.values, palette="viridis")

plt.title("In-Game Purchases by Income Range", fontsize=16)
plt.xlabel("Income Range", fontsize=14)
plt.ylabel("Count of In-Game Purchases", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df = fd[["SessionsPerWeek", "AvgSessionDurationMinutes", "EstimatedAnnualIncome", "EngagementLevel", "InGamePurchases"]]

label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))  ## Converting non-numeric data to numeric for co-relation generation
    label_encoders[column] = le


correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, fmt='.2f', annot_kws={"size": 8})
plt.title('Feature Correlation Heatmap')
plt.show()
```
