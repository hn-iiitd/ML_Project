```python
print(fd["InGamePurchases"].value_counts())

sns.set(style="whitegrid")

labels = ["No Purchases", "Purchases"]
data = fd["InGamePurchases"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data,
        labels=labels,
        autopct="%1.2f%%",
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05, 0.05])
plt.title("In-Game Purchases", fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
for col in ["Gender", "Location", "GameGenre", "EngagementLevel"]:
    print(f"\n{col} vs. In-Game Purchases:")
    print(pd.crosstab(fd[col], fd["InGamePurchases"]))

    # Get the frequency count of the current column
    col_freq = fd[col].value_counts()

    # Set the style of Seaborn
    sns.set(style="whitegrid")

    # Create a bar plot using Seaborn with the 'viridis' palette
    plt.figure(figsize=(8, 5))
    sns.barplot(x=col_freq.index, y=col_freq.values, palette='viridis')

    # Adding titles and labels
    plt.title(f'Frequency of {col} Distribution', fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Count', fontsize=14)

    # Simplifying x-axis labels if needed
    plt.xticks(ticks=range(len(col_freq.index)),
               labels=[str(level) for level in col_freq.index],
               rotation=0)  # Adjust rotation if necessary

    # Ensure a tight layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python


# Engagement Level Distribution
print("\nEngagement Level Distribution:")
print(fd["EngagementLevel"].value_counts())

# Plot Engagement Level Distribution
plt.figure(figsize=(8, 6))
engagement_level_freq = fd["EngagementLevel"].value_counts()

# Set Seaborn style
sns.set(style="whitegrid")

# Plot using Seaborn barplot
sns.barplot(x=engagement_level_freq.index, y=engagement_level_freq.values, palette="viridis")

# Add titles and labels
plt.title("Engagement Level Distribution", fontsize=16)
plt.xlabel("Engagement Level", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Plotting the frequency of In-Game Purchases by Engagement Level
purchases = fd[fd['InGamePurchases'] == 1]
print("\nEngagement Level Distribution for In-Game Purchases:")
print(purchases["EngagementLevel"].value_counts())

# Group by 'EngagementLevel' and count occurrences
engagement_level_freq = purchases.groupby('EngagementLevel').size()

# Create a bar plot for In-Game Purchases by Engagement Level
plt.figure(figsize=(8, 6))
sns.barplot(x=engagement_level_freq.index, y=engagement_level_freq.values, palette="magma")

# Add titles and labels
plt.title('Frequency of In-Game Purchases by Engagement Level', fontsize=16)
plt.xlabel('Engagement Level', fontsize=14)
plt.ylabel('Frequency of In-Game Purchases', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Pie chart: Distribution of Purchases by Engagement Level
grouped_data = fd.groupby("EngagementLevel")["InGamePurchases"].value_counts().unstack()
total_purchases_by_engagement = grouped_data.sum(axis=1)

# Create a pie chart using Seaborn's pastel palette
plt.figure(figsize=(6, 6))
plt.pie(total_purchases_by_engagement,
        labels=total_purchases_by_engagement.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("pastel"),
        explode=[0.05] * len(total_purchases_by_engagement))  # Slightly explode each slice

# Add title
plt.title("Purchases vs Engagement Level", fontsize=16)

# Ensure the pie is drawn as a circle
plt.axis('equal')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'fd' is your dataset

# Create Income Bins
fd["IncomeBin"] = pd.cut(fd["EstimatedAnnualIncome"], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000])

# Group by IncomeBin and count in-game purchases (only for '1')
grouped_data = fd.groupby("IncomeBin")["InGamePurchases"].apply(lambda x: (x == 1).sum())

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar chart using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=grouped_data.index.astype(str), y=grouped_data.values, palette="viridis")

# Add titles and labels
plt.title("In-Game Purchases by Income Range", fontsize=16)
plt.xlabel("Income Range", fontsize=14)
plt.ylabel("Count of In-Game Purchases", fontsize=14)
plt.xticks(rotation=45)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
